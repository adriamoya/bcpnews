```python
# Imports

import os
import json
import pprint
import datetime
import pandas as pd


# Variables

pp = pprint.PrettyPrinter(indent=4)

# Read data

data = []
with open('./articles_standar.json') as input_file:
    for line in input_file:
        data.append(json.loads(line))
```

```python
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
pp.pprint(list(df.columns))
```

```python
# Data Wrangling

keywords = []

for idx, row in df.iterrows():
    authors = ' / '.join(row['authors'])
    df.loc[idx, 'authors'] = authors
    
    for keyw in row['keywords']:
        keywords.append(keyw)
```

```python
df.groupby(['authors']).size().reset_index().rename(columns={0: "count"})
```

```python
# keywords analysis

df_kw = pd.DataFrame(pd.Series(keywords)).rename(columns={0: "keyword"})
```

```python
df_kw.groupby('keyword').size().sort_values(0, ascending=False).reset_index().rename(columns={0: "count"})
```
